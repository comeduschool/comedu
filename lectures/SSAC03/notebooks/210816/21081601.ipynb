### 포매팅의 활용
**{:[[fill]align][sign][#][0][width][,][.precision][type]}**

- fill: 채워지는 문자
- align: 문자열의 위치
  - ' '<: 문자열은 왼쪽으로 정렬시키고 빈칸으로 채움
- sign: 부호를 붙일 여부
  - '+': 양수와 음수에 부호를 붙임
  - '-': 음수에만 부호를 붙임 (기본값)
  - ' ': 양수에는 공백을 추가
- #: 정수를, 2진수, 8진수, 16진수로 나타낼때, 앞에 '0b', '0o', '0x'를 앞에 붙일지 나타냄.
- 0: 숫자에서 비어있는 공간은 0으로 채움
- ,: 천단위에 ','추가
- width: 포매팅할 문자열의 길이
- precision: 소수점 이하의 갯수
- type: 데이터의 타입
  - b: binary
  - o: octat
  - x: hex
  - d: decimal
  - f: float
  - c: character
  - s: string
  - %: percentage

- 문자열의 포매팅
  - [[fill]align][width]
  - align:
    - \>: right
    - <: left
    - ^: center

```python
"{0:*>10} and {1}".format("Python", "World")
```

```python
"{:<10}".format("Python")
```

```python
"{:^10}".format("Python")
```

```python
my_str = "Python"
```

```python
f"{my_str:*>10}"
```

```python
f"{my_str:<10}"
```

```python
f"{my_str:^10}"
```

# 반복문 응용

```python
eng = [99, 75, 35, 83, 97, 100]
math = [67, 78, 100, 82, 85, 68]
```

```python
for a in zip(eng, math):
  print(a)
```

```python
for e, m in zip(eng, math):
  print(e, m)
```

```python
passfail = []
for e, m in zip(eng, math):
  if e > 70 and m > 70:
    passfail.append("PASS")
  else:
    passfail.append("FAIL")
print(passfail)
```

```python
dict_one = {'name': 'Tony', 'last_name': 'Stark', 'job': 'Robot Programming'}
dict_two = {'name': 'Steven', 'last_name': 'Rogers', 'job': 'Combat Consultant'}

print(list(dict_one.items()))
print(list(dict_two.items()))
```

```python
for p1, p2 in zip(dict_one.items(), dict_two.items()):
  if p1[0] == 'name':
    print(p1[1], p2[1])
```
