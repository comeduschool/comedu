## json module
- JSON: 웹에서 데이터를 주고받는 형식
- dumps: dict -> json
- loads: json -> dict
- dump: dict -> json 파일
- loads: json파일 -> dict

```python
import json
```

```python
my_dict = {
    "a": 10,
    "b": 20,
    "c": 30,
    "d": 40
}
my_dict
```

```python
my_json_str = json.dumps(my_dict)
my_json_str
```

```python
my_dict1 = json.loads(my_json_str)
print(type(my_dict1))
my_dict1
```

```python
with open("scores.json", "w") as f:
  json.dump(my_dict, f, indent=4)
```

```python
with open("scores.json", "r") as f:
  my_dict2 = json.load(f)
print(my_dict2)
print(type(my_dict2))
```

## CSV (Comma Separted Values)

- writer
  - writerow
- reader
- DictReader
- DictWriter

```python
import csv
```

```python
with open("my_csv.csv", "w", encoding='euc-kr') as f:
  csv_writer = csv.writer(f)
  csv_writer.writerow(['이름', '출근시간', '퇴근시간', '시급', '근무시간', '일당'])
  csv_writer.writerow(['둘리', '5', '15', '8000', '10'])
  csv_writer.writerow(['고길동', '10', '22', '7800', '12'])
  csv_writer.writerow(['희동이', '15', '18', '9000', '3'])
```

```python
with open("my_csv.csv", "r", encoding="euc-kr") as f:
  csv_reader = csv.reader(f)
  for row in csv_reader:
    print(row)
```

```python
with open("my_csv.csv", "r", encoding="euc-kr") as f:
  dict_reader = csv.DictReader(f)
  for row in dict_reader:
    print(row)
    print(row['퇴근시간'])
```

```python
with open("my_csv1.csv", "w", encoding="euc-kr") as f:
  writer = csv.DictWriter(f, fieldnames=['이름', '출근시간', '퇴근시간', '시급', '근무시간','일당'])
  writer.writeheader()
  writer.writerow({
      "이름": "둘리",
      "퇴근시간": 15,
      "출근시간": 5,
      "시급": 8000,
      "근무시간": 10
  })
  writer.writerow({
      "이름": "고길동",
      "퇴근시간": 13,
      "출근시간": 15,
      "시급": 8000,
      "근무시간": 10
  })
  writer.writerow({
      "이름": "희동이",
      "퇴근시간": 22,
      "출근시간": 10,
      "시급": 8000,
      "근무시간": 10
  })
```
