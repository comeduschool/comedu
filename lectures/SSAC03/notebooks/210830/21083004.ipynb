## Map & Apply

```python
def fn(value):
  print(f"v->{value}")
  if value <= 300:
    return value + 20
  else:
    return value

def fn1(value):
  if value <= 300:
    return value + 20
  else:
    return value

def fn2(value):
  return value + 20 if value <= 300 else value
```

```python
my_ss = pd.Series([i*100 for i in range(1, 6)], index=list("abcde"))
my_ss
```

```python
my_result = my_ss.map(fn)
```

```python
my_result
```

```python
my_result1 = my_ss.map(fn2)
```

```python
my_result1
```

```python
my_ss.map(lambda v:v+20 if v<=300 else v)
```

```python
my_ss.apply(fn)
```
