## 반복
- *: 0개 이상
- +: 1개 이상
- {m,n}: m개 이상, n개 이하의 갯수
  - [a]{2,3}: aa, aaa
  - {n}: n개
  - {0,}: 0개 이상
    - a{0,}: "", "a", "aa", "aaa", ...
    - a*: "", "a", "aa", "aaa", ...
  - {1,}: 1개 이상
    - a{1,}: "a", "aa", "aaa", ...
    - a+: "a", "aa", "aaa", ...
  - {0,1}: 0개 또는 1개
    - a{0,1}: "", "a"
- ?: 있거나 없거나
  - a?: "", "a"

```python
re.search(".*", "a")
```

```python
re.search(".*", "1234dfsdf")
```

```python
re.search(".{4}", "124")
```

```python
re.search(".{4}", "1234")
```

```python
re.search("[0-9]{4,5}", "12345")
```

```python
# abcabcabc
re.findall("abc+", "abcabccabccc") # abc, abcc, abccc...
```

```python
re.findall("(abc)+", "abcaabcabcaabcabcabc") # abc, abcabc, abcabcabc
```

```python
re.findall("((abc)+)", "abcaabcabcaabcabcabc") # abc, abcabc, abcabcabc
# 8192389-2332333 -> 8192389-2******
```

```python
# abcabcabcabcabc
for matched in re.finditer("(abc)+", "abcaabcabcbabcabcbbabcabcabcabcabcabc"):
  print(matched.group())
```

- findall:
  - 그룹이 없으면, 일치하는 모든 문자열 
  - 그룹이 있으면:
    - 1개 그룹에 해당하는 문자열
    - 2개 그룹별 모든 문자열을 튜플로

## sub

```python
my_privates = """
  홍길동 900101-1011111
  고길동 801231-2022222
  이둘리 990326-1234567
  김둘리 011211-3234567
  박군인 800905-1049118
  김선생 070417-4059119
  김구라 070417-5059119
  박구라 070417-6059119
"""
```

```python
my_pat = re.compile("((\d{6}-[1|2|3|4])\d{6})")
```

```python
for matched in my_pat.finditer(my_privates):
  print(matched.group(2))
```

```python
print(my_pat.sub("PRIVATE", my_privates))
```

```python
print(my_pat.sub("\g<2>" + "*" * 6, my_privates))
```
