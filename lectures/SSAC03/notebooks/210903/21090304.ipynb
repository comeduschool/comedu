## 평균이 50점 이상이면 합격, 아니면 불합격인 Result 컬럼을 추가하시오.
- lambda
- function

```python
my_df.apply(lambda r: "합격" if r.mean()>=50 else "불합격", axis='columns')
```

```python
my_df['Result'] = my_df.apply(lambda r:"합격" if r.mean() >= 50 else "불합격", axis="columns")
my_df
```

```python
my_df = my_df.drop("Result", axis="columns")
my_df
```

```python
def my_pass(row):
  avg = row.mean()
  if avg >= 50:
    return "합격"
  else:
    return "불합격"
```

```python
my_df.apply(my_pass, axis="columns")
```

```python
my_df["Result"] = my_df.apply(my_pass, axis="columns")
my_df
```

## apply의 활용

```python
my_mean = my_df.kor.mean()
def fn(v):
  if v <= 40:
    return my_mean
  else:
    return v
```

```python
def fn(s):
  my_mean = s.mean()
  return s.apply(lambda v:my_mean if v<=40 else v)
```

```python
my_df.apply(fn)
```

```python
my_df.applymap(fn) # value by value
```
