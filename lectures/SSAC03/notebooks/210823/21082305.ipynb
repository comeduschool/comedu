# Heap
- Tree 중 하나
- 최소값/최대값으로 부모노드를 구성
- 가장 작은 값/ 가장 큰 값을 찾는데 사용

- 사용이유:
  - 정렬보다 빠름
  - 전체 정렬하지 않고도 최대/최소값을 찾을 수 있다.
  - **빠르게** 최대값과 최소값을 찾을 수 있다.

- heapify: 힙을 만든다.
- heappush(h, data): 힙에 데이터를 넣는다.
- heappop(h): 힙에서 데이터를 반환(최소값)
