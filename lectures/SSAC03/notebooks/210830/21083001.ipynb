## Series 생성시 전달 가능한 속성(Attribute)
- index
- name
- dtype
  - object
    - 2개 이상의 타입
    - String로만

```python
my_index
```

```python
len(my_list), len(my_index)
```

```python
my_ss = pd.Series(my_list, index=my_index)
my_ss
```

```python
my_ss = pd.Series(my_list, index=my_index)
my_ss
```

```python
my_ss = pd.Series(my_dict, index=["a", "bb", "cc", "dd"])
my_ss
```

```python
my_ss = pd.Series(my_list, dtype=np.str)
my_ss # "0", "1"
```

```python
my_ss = pd.Series(my_list, dtype=np.float64)
my_ss
```

```python
my_ss = pd.Series(my_list, name="kor")
my_ss
```
