# zip 함수

```python
eng = [99, 75, 35, 83, 97, 100]
math = [91, 87, 78, 66, 99, 82]
```

```python
for e, m in zip(eng, math):
  print(e, m)
```

```python
for i, e in enumerate(eng):
  for j, m in enumerate(math):
    if i == j:
      print(e, m)
      break
```

```python
passfail = []
for e, m in zip(eng, math):
  if e > 70 and m > 70:
    passfail.append("PASS")
  else:
    passfail.append("FAIL")
passfail
```

```python
dict_one = {"name": "Tony", "last_name": "Stark", "job": "Robot Programmer"}
dict_two = {"name": "Steve", "last_name": "Rogers", "job": "Combat Consultant"}
for p1, p2 in zip(dict_one.items(), dict_two.items()):
  print(p1[1], p2[1])
```

### Dictionary 축약형

```python
my_subjects = ["kor", "math", "eng"]
my_scores = [95, 25, 30]
```

```python
{i:sub for i, sub in enumerate(my_subjects)} # {key:value for i in range(n)}
```

```python
{sub: score for sub, score in zip(my_subjects, my_scores)}
```

```python
my_dict = {
    "a": 10,
    "b": 20,
    "c": 30,
    "d": 40
}
my_dict
```

```python
{k: my_dict[k] for k, v in my_dict.items()}
```

```python
{k+"_mid": my_dict[k] for k, v in my_dict.items()}
```

```python
my_dict2 = {
    "a": [98, 85, 75],
    "b": [82, 78, 99],
    "c": [100, 100, 98]
}
my_dict2
```
