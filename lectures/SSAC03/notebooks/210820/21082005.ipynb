## ZeroDivisionError

```python
my_num = 12
my_list = list(range(5))
for n in my_list[::]:
  try:
    value = my_num / n
  except ZeroDivisionError:
    value = None
  # print(value)
  if value is None: # 0으로 나눴다면, value가 무한수
    continue
  value = value * 3 # Blah, blah
  print(value)
```

## ValueError
- 함수에 전달된 값이 유효하지 않은 값일 때. <br>
(처리할 수 없는 값일 때)


```python
my_str = "12"
my_int = int(my_str) # int(my_str, base=10)
my_int
```

```python
my_str = "a"
my_int = int(my_str)
my_int
```

```python
import string
print(string.ascii_lowercase)
print(string.ascii_uppercase)
print(string.ascii_letters)
```

```python
my_str = "a"
try:
  my_int = int(my_str)
except ValueError:
  if my_str in string.ascii_letters:
    my_int = ord(my_str)
  else:
    my_int = None
print(my_int)
```

## 예외 처리
- try-except : try 실행시 에러가 발생하면 except 실행
- try-except(-else): try 실행시 에러가 발생하면 except 실행 발생하지 않으면 else 를 실행
- try-except(-else)-finally: try 실행시 에러가 발생해도 발생하지 않아도 finally 실행

```python
my_dict = {"a": 1}
try:
  my_dict["a"]
  print("Try")
except KeyError:
  print("Except")
else:
  print("Else: No Error")
```

```python
try:
  my_dict["b"]
  print("Try")
except KeyError:
  print("Except: Error")
else:
  print("Else: No Error")
```

```python
my_dict = {"a": 1}
try:
  my_dict["a"]
  print("Try")
except KeyError:
  print("Except: Error")
else:
  print("Else: No Error")
finally:
  print("Finally: Always")
```

```python
try:
  my_dict["b"]
  print("Try")
except KeyError:
  print("Except: Error")
else:
  print("Else: No Error")
finally:
  print("Finally: Always")
```
