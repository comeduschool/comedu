# heapify vs heappush

````python
my_list = [14, 2, 8, 12, -5, 17]
my_list
````

````python
my_list1 = my_list[:]
my_list1
````

````python
heapify(my_list1)
my_list1
````

````python
my_list2 = []
for num in my_list1:
  heappush(my_list2, num)
print(my_list2)
````

````python
my_list3 = [(-num, num) for num in my_list]
my_list3
````

````python
heapify(my_list3)
my_list3
````

````python
heappop(my_list3)
````

````python
heappop(my_list3)
````

````python
heappop(my_list3)[1]
````

## strftime vs strptime

- strftime : 문자열 -> 객체
- strptime : 객체 -> 문자열

### Format Code

- %a: 요일 축약형 (Sun, Mon, …, Sat)
- %A: 요일 전체이름 (Sunday, Monday, …, Saturday)
- %b: 월 이름 축약형 (Jan, Feb, …, Dec)
- %B: 월 이름 전체이름 (January, February, …, December )
- %Z: 시간대 (UTC, GMT)
- %j: 0으로 채워진 연중 일
- %y: 2자리 연도

- %d: 0으로 채워진 월중 일 (01, 02, …, 31)
- %m: 0으로 채워진 월
- %Y: 4자리 연도
- %H: 24시간 시간
- %I: 12시간 시간
- %M: 분
- %S: 초
