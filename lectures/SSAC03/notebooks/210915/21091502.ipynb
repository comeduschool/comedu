```python
values = np.linspace(-5, 5, 51)
result = lambda x: (1 / np.sqrt(2 * np.pi)) * np.exp(- x ** 2 / 2 )
my_df = pd.DataFrame({"values": values, "results": [result(x) for x in values]})
my_df
```

```python
my_df.plot.scatter(x='values', y='results')
```

```python
my_df.sample(5)
```

```python
my_df.sample(5).plot.scatter(x='values', y='results')
```

```python
my_df.sample(25).plot.scatter(x='values', y='results')
```

```python
my_df10 = my_df.sample(20)
my_df10.reindex(range(0, 51)).interpolate().plot.scatter(x='values', y='results')
```

```python
my_df10.reindex(range(0, 51)).interpolate(method='akima').plot.scatter(x='values', y='results')
```

```python
my_df['results2'] = my_df['results'].cumsum()
my_df.plot.scatter(x='values', y='results2')
```

```python
my_df20 = my_df.sample(15).reindex(range(0, 51))
my_df20
```

```python
my_df20.plot.scatter(x='values', y='results2')
```

```python
my_df20.interpolate(method='pchip').plot.scatter(x='values', y='results2')
```
