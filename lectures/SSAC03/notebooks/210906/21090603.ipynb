```python
import numpy as np
import pandas as pd
```

```python
# data
my_dict = {
  'kor': {
    'a': 10, 
    'b': 30,
    'c': 50,
    'd': 70,
    'e': 30,
  },
  'eng': {
    'a': 20, 
    'b': 40, 
    'c': 60, 
    'd': 80,
    'e': 100,
  },
  'math': {
    'a': 33, 
    'b': 44, 
    'c': 55, 
    'd': 66,
    'e': 77,
  },
}
```

```python
my_df = pd.DataFrame(my_dict)
my_df
```

# Data Manipulation
- assign
- where & mask
- apply, map, and applymap

## assign
- dict 표현식과 동일

```python
my_df.assign(total=lambda df:df.sum(axis='columns'))
```

```python
my_df['total'] = my_df.sum(axis='columns')
my_df
```

```python
my_df.assign(total=lambda df:df.sum(axis='columns'), avg=lambda df:df.mean(axis='columns'))
```

## where & mask
- 값 선택시:
  - where: 조건에 맞는 값 선택
  - mask: 조건에 맞지 않는 값 선택
- 값을 변경:
  - where: 조건에 맞지 않는 값을 변경
  - **mask**: 조건엔 맞는 값을 변경
    - nan
    - minus, '', ...

```python
my_df.where(my_df>50)
```

```python
my_df.mask(my_df>50)
```

```python
my_df.where(my_df>50, -100)
```

```python
my_df.mask(my_df>50, -100) #
```

```python
my_df.mask(my_df>50, my_df.mean(), axis='columns')
```

```python
my_df.mask(my_df>40, my_df.mean(axis='columns'), axis='index')
```

## apply, map, applymap
- map: python의 map을 구현
  - Series에만 구현
- apply: 
  - Series: 각 값에 적용
  - DataFrame: 각 열에 적용
- applymap:
  - DataFrame만 
  - 각 원소에 대해서


```python
list(map(lambda x:x+1, [1, 2, 3]))
```

```python
my_df.kor.map(lambda x:x+1)
```

```python
my_df.kor.apply(lambda x:x+1)
```

```python
my_df.apply(lambda s:s-s.mean())
```

```python
my_df.applymap(lambda x:x+10)
```

## 집계 (aggregation)
- reshape
  - (5, 3) => (10, 5)
- 행과 열, 그룹 특징(통계량)

```python
def my_mean(ss):
  return ss.mean() + 10
```

```python
my_df.agg([my_mean, np.mean])
```

```python
my_df.mean()
```

```python
my_df.median()
```

```python
my_df.describe().loc[['mean', 'count']]
```

```python
my_df.aggregate('mean')
```

```python
my_df.agg({'kor': np.mean, 'eng': np.median})
```
