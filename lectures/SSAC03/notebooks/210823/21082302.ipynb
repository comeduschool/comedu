## 예외 처리
- try-except
- try-except(-else)
- try-except(-else)-finally

```python
my_dict = {"a": 1}
try:
  my_dict["a"]
  print("Try: After error.")
except KeyError:
  print("Except: Error Raised.")
```

```python
try:
  my_dict["b"]
  print("Try: After error.")
except KeyError:
  print("Except: Error Raised.")
```

```python
try:
  my_dict["a"]
  print("Try: After error.")
except KeyError:
  print("Except: Error Raised.")
else:
  print("Else: No error")
  
```

```python
try:
  my_dict["a"]
  print("Try: After error.")
except KeyError:
  print("Except: Error Raised.")
else:
  print("Else: No error.")
finally:
  print("Finally: Error & No Error")
```

```python
try:
  my_dict["b"]
  print("Try: After error.")
except KeyError:
  print("Except: Error Raised.")
else: 
  print("Else: No Error.")
finally:
  print("Finally: Error & No Error.")
```

try실행
- -> 에러발생 (O) catch 실행 + finally 실행
- -> 에러발생 (X) else 실행 + finally 실행

```python
try:
  my_dict["b"]
except KeyError:
  print("Handle Key Error")
except IndexError:
  print("Handle Index Error")
```

```python
my_list = [1, 2, 3]
```

```python
try:
  my_list[5]
except KeyError:
  print("Handle Key Error")
except IndexError:
  print("Handle Index Error")
except Exception:
  print("Other Exception")
```

```python
try:
  my_dict["b"]
except (KeyError, IndexError):
  print("Both Error")
```

```python
try:
  my_dict["b"]
except Exception:
  print("All Exception")
```
