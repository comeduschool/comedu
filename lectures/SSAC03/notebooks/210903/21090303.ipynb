## 라벨(인덱스)의 인덱싱

  - my_df.loc[Label]
  - my_df.loc[[Label]]
  
- 여러행 가져오기
  - my_df.loc[[Label1, Label2]]

```python
my_df['kor']
```

```python
my_df.loc['a']
```

```python
my_df.loc[['a']]
```

```python
my_df.loc[['a', 'c']]
```

원소의 인덱싱:
- my_df[COLUMN_LABEL][INDEX_LABEL]
  - my_df[COLUMN_LABEL].loc[INDEX_LABEL]
- my_df.loc[INDEX_LABEL, COLUMN_LABEL]
