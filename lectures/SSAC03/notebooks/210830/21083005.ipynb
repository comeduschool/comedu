# DataFrame data types
Pandas의 데이터 타입들은 기본적인 파이썬 타입이나 numpy의 타입들로 생성할 수 있다.


|result  |types                    |
|--------|-------------------------|
|'b'     | boolean                 |
|'i'     | (signed) integer        |
|'u'     | unsigned integer        |
|'f'     | floating-point          |
|'c'     | complex-floating point  |
|'O'     | (Python) objects        |
|'S', 'a'| (byte-)string           |
|'U'     | Unicode                 |
|'V'     | raw data (void)         |
