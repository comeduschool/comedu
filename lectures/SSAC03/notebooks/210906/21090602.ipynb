```python
import numpy as np
import pandas as pd
```

```python
# data
my_dict = {
  'kor': {
    'a': 10, 
    'b': 30,
    'c': 50,
    'd': 70
  },
  'eng': {
    'a': 20, 
    'b': 40, 
    'c': 60, 
    'd': 80
  },
  'math': {
    'a': 33, 
    'b': 44, 
    'c': 55, 
    'd': 66
  },
}
```

```python
my_df = pd.DataFrame(my_dict)
```

# Data Scenario
- 첫번째 값
- 마지막 값
- 3번째 값
- (0, 3)의 값

- a의 값
- a와 d의 값
- ('a', 'math')의 값

- 2번째부터 3번째 값까지
- 3번째 값까지
- 3번쨰 값부터 마지막 값까지

- **a와 d 사이의 값** -> 인덱스(라벨)
  - '2020. 2월' 부터 '2020. 3월' 

- a행의 값
- **math열의 값** -> 열
  - '미국'의 확진자 데이터


# DataFrame: Data Selection
- 라벨로 값 선택하기
  - dict
  - .
  - loc
- 위치로 값 선택하기
  - iloc
- 리스트로 값 선택하기

```python
my_df.kor
```

```python
my_df['kor']
```

```python
my_df.loc['a']
```

```python
my_df.loc[:, 'kor']
```

```python
my_df.iloc[0]
```

```python
my_df.iloc[-1]
```

```python
my_df[['kor', 'eng']] # 특수 케이스
```

## 슬라이싱
