### 날짜시간의 포매팅
- %a: 요일 축약형 (Sun, Mon, ... Sat)
- %A: 요일 (Sunday...)
- %d: 월중 일 (01, 02, ... 31)
- %b: 월 이름 축약형 (Jan, Feb... Dec)
- %B: 월 이름 (January, ..)
- %m: 월 (01, 02, 12)
- %y: 연도 (2자리)
- %Y: 연도 (4자리)
- %H: 24시간 (01, 24)
- %I: 12시간 (01, 12)
- %M: 분
- %S: 초
- %j: 연중일 (1,.. 365)  (3월 5일은 몇번째 날인가?)

#### 꼭 알아야할 것

- %d: 월중 일 (01, 02, ... 31)
- %m: 월 (01, 02, 12)
- %y: 연도 (2자리)
- %Y: 연도 (4자리)
- %H: 24시간 (01, 24)
- %I: 12시간 (01, 12)
- %M: 분
- %S: 초

```python
from datetime import datetime
```

```python
today = datetime.now()
```

```python
today # 기준시
```

```python
"{:%Y-%m-%d %I:%M:%S}".format(today)
```

```python
"{:%d/%m/%Y %I:%M:%S}".format(today)
```

```python
f"{today:%Y-%m-%d %I:%M}"
```

```python
today.month
```

```python
f"{today:%m}"
```

```python
f"{today:%Y-%m-%d (%a) %I:%M}"
```

```python
int(f"{today:%j}") % 7 # + 1월 1일의 요일
```

```python
int(f"{today:%w}") # Fri
```
