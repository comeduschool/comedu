## Attribute

# **비교 연산**
- 일치
- not 일치
- 크다/크거나 같다
- 작다/작거나 같다
- 포함(**isin**)

```python
my_ss1 == 200
```

```python
my_ss1[my_ss1==200]
```

```python
my_ss1 > 200
```

```python
my_ss1[my_ss1>200]
```

```python
my_ss1[[True, False, True, False, True]]
```

```python
my_ss1.isin([200, 100])
```

```python
my_ss1[my_ss1.isin([100, 200])]
```

# 논리 연산
- AND (&) 
- OR (|)
- NOT (~)

```python
my_ss1 > 100
```

```python
my_ss1 <= 300
```

```python
(my_ss1 > 100) & (my_ss1 <= 200)
```

```python
my_ss1[(my_ss1 >100) & (my_ss1 <= 200)]
```

```python
my_ss1 > 100
```

```python
my_ss1[my_ss1>100]
```

```python
my_ss1[~(my_ss1>100)]
```

# or vs isin

```python
my_ss1 < 200
```

```python
my_ss1 > 300
```

```python
my_ss1[(my_ss1 < 200) | (my_ss1 > 300)]
```

```python
my_ss1[my_ss1==100]
```

```python
my_ss1[my_ss1==200]
```

```python
my_ss1[(my_ss1==100)|(my_ss1==200)|(my_ss1==300)|(my_ss1==400)]
```

```python
my_ss1[my_ss1.isin([100, 200, 300, 400])]
```
