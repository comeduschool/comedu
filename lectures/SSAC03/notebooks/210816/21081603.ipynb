## 매개변수의 정의
- 위치 매개변수 (Positional Parameter): 필수일때

- 이름을 가지는 매개변수
- 가변 매개변수

  - 이름을 가지는 매개변수

````python
def my_fn(a, b, c):
  return a*b+c
````

````python
def my_fn9(*args, **kwargs):
  print(args)
  print(kwargs)
````

````python
my_fn9(1, 2, 3, x=1, y=2, z=3)
````

````python
def my_fn10(start, *args):
  if args:
    return start + sum(args)
  else:
    return start
````

````python
my_fn10(10)
````

````python
my_fn10(1, 2, 3)
````

````python
def my_fn11(start, *args):
  if not args:
    return start
  else:
    return start + sum(args)
````

````python
my_fn11(10)
````

````python
my_fn11(1, 2)
````

````python
def my_fn12(num, **kwargs):
  print(kwargs)
  if 'base' in kwargs:
    base = kwargs['base']
    return f"{num:{base}}" # {100:b} {100:x}
  else:
    return num
````

````python
my_fn12(10) # 10진수
````

````python
my_fn12(10, base='b') # 2진수
````

````python
my_fn12(10, base='x') # 16진수
````

### 매개변수 펼치기 (Unfold)
- list/tuple 매개변수 펼치기
- 사전 매개변수 펼치기

````python
my_scores1 = [92, 88, 76]
my_scores1
````

````python
my_scores2 = {
    'a': 92,
    'b': 88,
    'c': 92
}
my_scores2
````

````python
def my_fn13(a, b, c):
  print(a)
  print(b)
  print(c)
````

````python
my_fn13(*my_scores1) # == my_fn13(92, 88, 76), 길이가 중요
````

````python
def my_fn14(*args):
  print(args)
````

````python
def my_fn15(a):
  print(a)
  a[0] = 100
````

````python
my_fn14(my_scores1)
````

````python
my_fn14(*my_scores1)
````

````python
my_fn15(my_scores1)
````

````python
my_fn15(*my_scores1)
````

````python
print(my_scores1)
my_fn15(my_scores1)
print(my_scores1)
````

````python
def my_fn16(c, b, a):
  print(a)
  print(b)
  print(c)
````

````python
my_fn16(**my_scores2) # == my_fn16(a=92, b=88, c=92)
````

````python
def my_fn17(a, **kwargs):
  print(a)
  print(kwargs)
````

````python
my_fn17(**my_scores2)
````

````python
def my_fn18(**kwargs):
  print(kwargs)
````

````python
my_fn18(**my_scores2) # == my_fn18(a=92, b=88, c=92)
````

## 일급 클래스(First Class)로써 함수

- 함수의 반환값으로 사용 가능

### 변수에 저장 가능

````python
def my_fn19():
  return "Hello"
````

````python
a = my_fn19
````

````python
a
````

````python
a()
````

### 매개변수로 전달 가능

````python
def my_fn20(fn):
  return fn()
````

````python
my_fn20(my_fn19)
````

### 함수의 반환값으로 사용 가능

````python
def my_fn21(fn):
  return fn
````

````python
a = my_fn21(my_fn19)
````
