## 최대값 최소값

### 최대값
- max(iter, *[, key, default])
- max(arg1, arg2, *args[, key])

### 최소값
- min(iter, *[, key, default])
- min(arg1, arg2, *args[, key]) 

```python
max(2, 3)
```

```python
max(7, 5, 9)
```

```python
max([7, 5, 9]) # iterable object, 반복 가능한 객체
```

```python
max([], default=100)
```

```python
min(2, 3)
```

```python
min(7, 5, 9)
```

```python
min([7, 5, 9])
```

```python
min([], default=100)
```

```python
blackpink = [
  {
    "name": "지수",
    "age": 25,
    "height": 162
  },
  {
    "name": "제니",
    "age": 24,
    "height": 163
  },
  {
    "name": "로제",
    "age": 23,
    "height": 168
  },
  {
    "name": "리사",
    "age": 24,
    "height": 166.5
  },
]
```

```python
max(blackpink, key=lambda m:m["age"])
```

```python
max(blackpink, key=lambda m:m["height"])
```

```python
min(blackpink, key=lambda m:m["age"])
```

```python
min(blackpink, key=lambda m:m["height"])
```

```python
min(blackpink, key=lambda x:x["height"])
```

### lambda (익명 함수)

```python
def func(a, b):
  return b, a
```

```python
func(2, 1)
```

```python
lambda a, b: b, a
```

```python
def height_key(member):
  return member["height"]
```

```python
min(blackpink, key=height_key)
```
