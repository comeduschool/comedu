### 가변 매개변수
- \* : 위치 가변 매개변수
  - args
- ** : 키워드 가변 매개변수
  - kwargs

````python
def my_fn2(num, *args):
  print(args)
````

````python
my_fn2(1, 2, 3, 4)
````

````python
def my_fn3(num, *nums):
  print("num = ", num)
  for n in nums:
    print(n)
````

````python
my_fn3(100, 1, 2, 3)
````

````python
def my_fn4(num, *nums):
  if nums:
    return num + nums[0]
  else:
    return num
````

````python
my_fn4(5, 10, 11)
````

## Unfold (Unfolding), 매개변수 펼치기
- a, b, c => [a, b, c]
- a=1, b=2, c=3 => {'a':1, 'b':2, 'c': 3}

````python
def my_fn14(a, b, c):
  return (a*b)%c
````

````python
my_list = [3, 7, 4]
````

````python
my_fn14(my_list[0], my_list[1], my_list[2])
````

````python
my_fn14(*my_list)
````

````python
my_list1 = [3, 7, 4, -100]
````

````python
my_fn14(*my_list1)
````

````python
def my_fn15(a, b, c, *args):
  print(args)
  if args:
    return a*b*c*args[0]
  else:
    return a*b*c
````

````python
my_fn15(*my_list) # == my_fn15(my_list[0], my_list[1], my_list[2])
````

````python
my_fn15(*my_list1) # my_fn15(my_list1[0], my_list1[1], my_list1[2], my_list1[3])
````

````python
my_fn15(*my_list1[:3])
````

````python
my_dict = {
    "a": 10,
    "b": 20,
    "c": 30
}
````

````python
my_fn14(a=my_dict["a"], b=my_dict["b"], c=my_dict["c"])
````

````python
my_fn14(**my_dict)
````

# 가변 위치 매개변수 vs 리스트, 튜플

````python
def my_sum(*args, start):
  init = 0
  for num in args:
    init += num
  return init
````

````python
my_sum(1, 2, 3, 4, 5)
````

````python
def my_sum1(nums, start=0):
  init = start
  for num in nums:
    init += num
  return init
````

````python
my_sum1([1, 2, 3, 4, 5])
````

````python
my_sum1([-1, -2, -3, -4, -5])
````

````python
my_sum1([-1, -2, -3, -4, -5], -1000)
````
