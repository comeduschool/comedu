## where vs mask
- where:
- mask

```python
my_df = pd.DataFrame(my_dict)
my_df
```

```python
my_df.where(my_df>50)
```

```python
my_df.mask(my_df>50)
```

```python
my_df.where(my_df<50, -100)
```

```python
my_df.mask(my_df<50, -100)
```

```python
my_df.kor.iloc[:2] = np.nan
my_df
```

```python
my_df.kor.iloc[:2] = np.nan
my_df
```

```python
my_df.where(my_df==-100)
```

```python
my_df.mean(axis='columns')
```

```python
my_df.mean()
```

```python
my_df.mask(my_df.kor.isna(), my_df.mean(), axis='columns')
```

## agg

```python
my_df.agg(np.mean)
```

```python
my_df.agg('sum')
```

```python
my_df.sum()
```

```python
my_df.agg(['mean', 'sum'])
```

```python
my_df.describe().loc[['count', 'mean']]
```

## map, apply, applymap
- apply
  - Series: 각 데이터별
  - DataFrame: 각 시리즈별

- map: Series
  - 파이썬 동작: iloc, map

- applymap: DataFrame
  - 각 원소의 접근 
  - apply -> apply

```python
my_df.apply(lambda v:v['kor']+v['eng'], axis="columns")
```
