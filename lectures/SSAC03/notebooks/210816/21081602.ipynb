```python
board = [[0,0,0,0,0],[0,0,1,0,3],[0,2,5,0,1],[4,2,4,4,2],[3,5,1,3,1]]
moves = [1,5,3,5,1,2,1,4]
my_cm2 = ChainMap(*[{idx+1: num for idx, num in enumerate(line) if num!=0} for line in board])
my_cm2
```

```python
poped = []
cnt = 0
for move in moves:
  for index, my_map in enumerate(my_cm2.maps):
    try:
      res = my_map[move]
      doll = my_cm2.maps[index].pop(move)
      if len(poped) == 0:
        poped.append(doll)
      else:
        if poped[-1] == doll:
          poped.pop()
          cnt = cnt + 2
        else:
          poped.append(doll)
      break
    except:
      continue
print(cnt) # 6
print(poped) # 9
```

```python
# https://docs.python.org/ko/3/library/functools.html
my_trim_mean = partial(stats.trim_mean, proportiontocut=0.1)
my_df.apply(my_trim_mean)
```
