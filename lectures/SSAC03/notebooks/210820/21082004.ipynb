```python
from functools import partial
```

```python
f"{1000:b}"
```

```python
int(f"{1000:b}", base=2)
```

```python
my_basetwo = partial(int, base=2)
my_basetwo
```

```python
my_basetwo(f"{1000:b}")
```

```python
pow(2, 3) # 2 ** 3
```

```python
my_twopow = partial(pow, 2)
my_twopow
```

```python
my_twopow(3) # 2 ** 3
```

```python
my_twopow(4) # 2 ** 4 
```

```python
my_twopow(5) # 2 ** 5
```
