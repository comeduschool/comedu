## Factorial
- 0! = 1
- 1! = 1
- n! = n(n-1)!

```python
# n! = n*n-1...* 1
```

```python
def factorial(n):
  if n == 0:
    return 1
  elif n == 1:
    return 1
  return n * factorial(n-1)
```

```python
factorial(0)
```

```python
factorial(1)
```

```python
factorial(5)
```

```python
factorial(10000)
```

# 피보나치
- fibo(1) = 1
- fibo(2) = 2
- fibo(n) = fibo(n-1) + fibo(n-2)

```python
def fibo(n):
  if n == 1:
    return 1
  elif n == 2:
    return 2
  return fibo(n-1) + fibo(n-2)
```

```python
fibo(1)
```

```python
fibo(5)
```

```python
[fibo(n) for n in range(1, 10)]
```

```python
memo = {1:1, 2:1}
def fibo1(n):
  global memo
  if memo.get(n) is None:
    memo[n] = fibo1(n-1) + fibo1(n-2)
  return memo[n]
```

```python
fibo1(5)
fibo1(4) fibo1(3)
```
