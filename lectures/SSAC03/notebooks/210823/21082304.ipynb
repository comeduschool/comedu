### strptime vs strftime
- fromisoformat(str) -> datetime

### Format Code

- %a: 요일 축약형 (Sun, Mon, …, Sat)
- %A: 요일 전체이름 (Sunday, Monday, …, Saturday)
- %d: 0으로 채워진 월중 일 (01, 02, …, 31)
- %b: 월 이름 축약형 (Jan, Feb, …, Dec)
- %B: 월 이름 전체이름 (January, February, …, December )
- %m: 0으로 채워진 월
- %y: 2자리 연도
- %Y: 4자리 연도
- %H: 24시간 시간
- %I: 12시간 시간
- %M: 분
- %S: 초
- %p: AM/PM
- %Z: 시간대 (UTC, GMT)
- %j: 0으로 채워진 연중 일
