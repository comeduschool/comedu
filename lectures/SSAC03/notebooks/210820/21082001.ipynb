## ChainMap
- 여러개의 dict <br>
 {"a":1, "c":2}, {"b": 3, "c":4}


```python
my_dict = {
    "a": 1,
    "c": 2
}
```

```python
my_dict1 = {
    "b": 3,
    "c": 4
}
```

```python
from collections import ChainMap
```

```python
my_cm = ChainMap(my_dict, my_dict1)
my_cm
```

```python
my_cm.maps
```

```python
my_cm.maps[0]["b"]
```

```python
my_cm.maps[1]["b"]
```

```python
list(my_cm.keys())
```

```python
list(my_cm.values())
```

```python
my_cm.parents
```

```python
my_cm.parents.maps
```

```python
my_cm.parents.parents
```

```python
"원소 있음" if my_cm.parents.parents else "원소 없음"
```

```python
my_cm = ChainMap(my_dict, my_dict1, my_dict.copy())
my_cm
```

```python
my_cm.parents
```

## Counter

```python
my_str = "aaabbbabbccdd"
```

```python
my_list = list(my_str)
my_list
```

```python
from collections import Counter
```

```python
Counter(my_str)
```

```python
set(my_str)
```

```python
my_counter = {}
for c in set(my_str):
  my_counter[c] = my_str.count(c)
my_counter
```

```python
my_counter1 = {}
for c in my_list:
  my_counter1[c] = my_counter1.get(c, 0) + 1
my_counter1
```

```python
my_counter2 = Counter(my_str)
```

```python
my_counter2.update({'z': 100})
my_counter2
```

```python
my_counter2.update(['z'])
my_counter2
```

```python
my_counter2.update(['z', 'z', 'z'])
my_counter2
```

```python
list(my_counter2.elements())
```

```python
my_counter2.most_common(1)
```

```python
my_counter2.most_common(1)[0][0]
```
