```python
from fake_useragent import UserAgent
ua = UserAgent()
```

```python
headers = {
    "User-Agent": ua.random
}
headers
```

```python
res = requests.get(url, headers=headers)
res
```

```python
res.text
```

```python
my_soup = BeautifulSoup(res.text, 'html.parser')
my_soup
```

```python
article_lists = my_soup.find_all('div', {'class': 'mtype_list_wide'})
article_lists[-1]
```

```python
articles = article_lists[-1].find_all('a')
articles
```

```python
for article_elem in articles:
  print("content link ->", article_elem['href'])
  print("title ->", article_elem.find('strong').text)
```
