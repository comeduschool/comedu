- findall
- finditer

```python
re.search("P.{5}", "PythonPandas")
```

```python
re.findall("P.{5}", "PythonPandaspaypal")
```

```python
for matched in re.finditer("P.{5}", "PythonPandaspaypal"):
  print(matched.group())
  print(matched.span())
```

```python
re.findall("P(.{5})", "PythonPandaspaypal") # capture
```

```python
re.findall("(P(.{5}))", "PythonPandaspaypal")
```

```python
for matched in re.finditer("(P(.{5}))", "PythonPandaspaypal"):
  print(matched.group(2))
```

## 반복:
- *: 0개 이상 반복
- +: 1개 이상 반복
- {m, n}: m개이상 n개이하
  - {m}: m개
  - {0,}: *
  - {1,}: +
  - {3,}: 3개이상
- ?: 있거나 없거나
  - {0, 1}
