## 슬라이싱
- 순서:
  - iloc (파이썬의 슬라이싱과 동일) 
  - iloc[start:stop:step]: [start, stop)
- 라벨을 이용한 슬라이싱:
  - loc
    - my_ss.loc[start:stop]: [start, stop]
  - dict 형식
    - my_ss[start:stop]: [start, stop]

```python
my_ss = pd.Series(my_list)
my_ss
```

```python
my_ss.iloc[2:] # n번째 부터 끝까지
```

```python
my_ss.iloc[:2]
```

```python
my_ss.iloc[2:4]
```

```python
my_ss.iloc[:]
```

```python
my_ss[:3] # 안티패턴
```
