## astype('category') vs dtype='category'
- 문자열 타입 -> 행의 갯수 <<<< unique() 원소의 갯수
- 숫자인 경우 -> 0-100: 
  - 100-90: 우수 
  - 90-0: 일반 
- encoding: onehot, 숫자

```python
my_ss = pd.Series(np.random.randint(0, 100, 100))
my_ss
```

```python
my_ss2 = pd.cut(my_ss, bins=[-1, 60, 70, 80, 90, 100], labels=['가', '양', '미', '우', '수'])
my_ss2
```

```python
my_ss2.cat.as_unordered().cat.as_ordered()
```

## range: inclusive vs exclusive

- (0, 10] : 0 < x <= 10
- [0, 10) : 0 <= x < 10

# Renaming Categories

```python
my_ss2 = my_df.B.astype('category')
my_ss2
```

```python
my_index = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']
my_categorical = pd.Categorical(['a', 'a', 'a', 'a', 'a', 'a', 'a', 'a'], categories=['a', 'b'])
my_nums = [1, 2, 3, 1, 1, 1, 1, 1]
my_df2 = pd.DataFrame({"values":my_nums, "category":my_categorical}, index=my_index)
my_df2
```
