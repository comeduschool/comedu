# csv module

csv: comma separted values

- writer
- reader
- DictWriter
- DictReader

````python
import csv
````

````python
with open("my_csv.csv", "w") as f:
  writer = csv.writer(f)
  writer.writerow(["이름", "출근시간", "퇴근시간", "시급", "근무시간", "일당"])
  writer.writerow(["둘리", "5", "15", "8000", "10"])
  writer.writerow(["고길동", "8", "22", "7800", "14"])
  writer.writerow(["희동이", "15", "18", "9000", "3"])
````

````python
with open("my_csv.csv", "r") as f:
  reader = csv.reader(f)
  for row in reader:
    print(row)
````

````python
with open("my_csv.csv", "r") as f:
  reader = csv.DictReader(f)
  for row in reader:
    print(row)
    print(row['이름'])
````
